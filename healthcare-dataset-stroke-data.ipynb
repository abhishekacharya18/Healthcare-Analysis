```python
import pandas as pd

# Load dataset
df = pd.read_csv(r'C:\Users\abhis\Downloads\Data_Analytics_Projects\Data Analysis\01_raw_data\healthcare-dataset-stroke-data.csv')
```

```python
df.head()
```

```python
df['gender'].unique()
```

```python
df = df[df['gender'] != 'Other']
```

```python
df.isnull().sum()
```

```python
df['bmi'] = df['bmi'].fillna(df['bmi'].mean())
```

```python
df = df.drop_duplicates()
```

```python
df.dtypes
```

```python
df['age'] = df['age'].astype(int)
```

```python
df['gender'].unique()
```

```python
df = df[df['gender'] != 'Other']
```

```python
df['bmi'].describe()
```

```python
df = df[(df['bmi'] > 10) & (df['bmi'] < 60)]
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.isnull().sum()
df.shape
```

```python
df['Gender'] = df['gender_Male'].apply(lambda x: 'Male' if x == 1 else 'Female')
```

```python
def age_group(age):
    if age < 30:
        return "Under 30"
    elif age < 50:
        return "30-50"
    elif age < 70:
        return "50-70"
    else:
        return "70+"

df['Age Group'] = df['age'].apply(age_group)
```
